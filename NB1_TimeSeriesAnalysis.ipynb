```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

qfilename = 'my_streamflow_data.pkl'
```

```python
df = pd.read_pickle(qfilename)
df
```

```python
df.dtypes
```

```python
Q_cms = df['Discharge']/(3.2808**3)
Q_cms
```

```python
df['Discharge_cms'] = Q_cms
df
```

```python
df.rename(columns={'Discharge':'Discharge_cfs'}, inplace=True)
df
```

```python
df.plot(x='Dates',y='Discharge_cms', figsize=(14,10))
```

```python
df['pd_dates'] = pd.to_datetime(df['Dates'])      # convert the numpy dates into pandas numbers

df
```

```python
df['month'] = df['Dates'].dt.month
df.dtypes
df
```

```python
df
```

```python
df.index = pd.to_datetime(df['pd_dates'], '%y-%m-%d')
df.groupby(by=df.index.month).mean()
```

```python
q_seasonal=df['Discharge_cms'].groupby(by = df.index.month).mean().values

ss = df['Discharge_cms'].groupby(df['month']).mean()
q_seasonal
```
